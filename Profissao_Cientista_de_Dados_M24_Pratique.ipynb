# **MÓDULO 24 - Bancos de Dados + SQL**

Para nossa atividade inicial de SQL trabalharemos com uma base de vendas.
Abaixo temos o ambiente de preparo para trabalharmos com o SQL lite.

```python
import sqlite3 #Importando a biblioteca necessária
import pandas as pd
from google.colab import files
import io
```

```python
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
df = pd.read_csv(io.BytesIO(uploaded[file_name]), delimiter=';')
print(df.head())
```

```python
conn = sqlite3.connect(':memory:')
df.to_sql('tb_vendas', conn, index=False, if_exists='replace')
```

```python
def run_query(query):
    return pd.read_sql_query(query, conn)
```

# 1) Faça uma consulta que retorne todos os dados da nossa tabela de vendas, com todas linhas e todas colunas.


```python
query_1 = "SELECT * FROM tb_vendas"
resultado = run_query(query_1)
resultado
```

# 2) Faça uma consulta que retorne apenas 10 linhas da coluna produto.


```python
query_2 = "SELECT PRODUTO FROM tb_vendas LIMIT 10"
resultado_2 = run_query(query_2)
resultado_2
```

# 3) Faça uma consulta que retorne a média do valor por unidade e também das unidades vendidas. Não se esqueça de renomear o nome das colunas ao cria-las.

```python
query_3 = """
SELECT
    AVG(VALOR_UNID) AS Media_Valor_Unitario,
    AVG(UNIDADES) AS Media_Unidades_Vendidas
FROM tb_vendas
"""

resultado_3 = run_query(query_3)
resultado_3
```

# 4) Note que na nossa tabela não temos o total de valor gasto por compra, temos o valor do produto e a quantidade vendida, para termos o valor total precisamos multiplicar essas duas colunas e para multiplicar colunas usamos o *.
# Faça uma consulta que retorne a coluna ID_COMPRA, ID_CLIENTE e o valor total gasto.

Ex: SELECT TICKET_MEDIO*UNIDADE FROM TB_EXEMPLO

```python
query_4 = """
SELECT
    ID_COMPRA,
    ID_CLIENTE,
    (VALOR_UNID * UNIDADES) AS VALOR_TOTAL_GASTO
FROM tb_vendas
"""

resultado_4 = run_query(query_4)
resultado_4
```

# 5) [DESAFIO] Faça uma consulta que traz a média do valor total gasto que foi calculado no exercício anterior. Note que não queremos a média por pedido e sim a média total, apenas essa informação deve vir na consulta.

```python
query_5 = """
SELECT
    AVG(VALOR_UNID * UNIDADES) AS MEDIA_TOTAL_GASTO
FROM tb_vendas
"""

resultado_5 = run_query(query_5)
resultado_5
```

#7) Diferencie com suas palavras banco de dados relacionais de não relacionais.


Bancos de dados relacionais organizam os dados em tabelas rígidas com linhas e colunas (como o SQL que usamos), sendo ideais para transações precisas e estruturadas. Já os não relacionais (NoSQL) oferecem flexibilidade total, armazenando dados em formatos variados como documentos ou grafos, sendo perfeitos para grandes volumes de dados desestruturados e rápida escalabilidade.

#8) Queremos saber da sua opinião, acredita que o SQL pode ser mais prático que o Python para consultas rápidas? Justifique.



Com certeza, o SQL costuma ser mais prático para consultas rápidas. Os motivos principais são:

-Proximidade com os Dados: O SQL trabalha diretamente no banco de dados. Você não precisa "trazer" os dados para o seu computador para analisá-los, o que economiza tempo e recursos.

-Sintaxe Direta (Declarativa): No SQL, você diz apenas o que quer (ex: "me dê a média de vendas"). No Python, você precisa dizer como fazer (importar bibliotecas, ler o arquivo, definir separadores, etc.).

-Performance com Grandes Volumes: Bancos de dados são otimizados para filtrar milhões de linhas em segundos. O Python (Pandas) pode travar ou ficar lento se o volume de dados for maior que a memória RAM do seu computador.

-Menos Código: Uma consulta que resolve um problema em uma linha de SQL pode exigir cinco ou seis linhas de código em Python.

Em resumo:

Uso SQL para extrair, filtrar e somar dados rapidamente. Uso Python para limpar dados complexos, criar gráficos ou fazer previsões (Machine Learning).
