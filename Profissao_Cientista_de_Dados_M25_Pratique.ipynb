# **MÓDULO 25 - Operações Básicas SQL**

Para nossa atividade inicial de SQL trabalharemos com uma base de vendas.
Abaixo temos o ambiente de preparo para trabalharmos com o SQL lite.

```python
import sqlite3 #Importando a biblioteca necessária
import pandas as pd
from google.colab import files
import io
```

```python
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
df = pd.read_csv(io.BytesIO(uploaded[file_name]), delimiter=';')
print(df.head())
```

```python
conn = sqlite3.connect(':memory:')
df.to_sql('tb_vendas', conn, index=False, if_exists='replace')
```

```python
def run_query(query):
    return pd.read_sql_query(query, conn)
```

# 1) Faça uma consulta que retorne os nomes dos produtos distintos que temos na base de venda.


```python
query_produtos = "SELECT DISTINCT PRODUTO FROM tb_vendas"


df_produtos_distintos = run_query(query_produtos)
print(df_produtos_distintos)
```

# 2) Faça uma consulta que retorne a contagem dos clientes distintos que temos na nossa base, não esqueça de renomear sua coluna.

Dica: Você pode usar o distinct dentro do count()

```python
query_contagem_clientes = """
SELECT COUNT(DISTINCT ID_CLIENTE) AS total_clientes_unicos
FROM tb_vendas
"""

resultado = run_query(query_contagem_clientes)
print(resultado)
```

# 3) Faça uma consulta que retorne uma coluna com os produtos distintos e o valor_unid de cada produto. Porém apenas para produtos onde o valor_unid é maior ou igual a 50 reais.

```python
query_produtos_caros = """
SELECT DISTINCT PRODUTO, VALOR_UNID
FROM tb_vendas
WHERE VALOR_UNID >= 50
"""

resultado = run_query(query_produtos_caros)
print(resultado)
```

# [DESAFIO] 4) Faça uma consulta que retorne o ID das compras e o valor total gasto nessa compra (Valor_unidade X Unidades) das 5 compras com mais valor total gasto.

```python
query_top_compras = """
SELECT
    ID_COMPRA,
    (CAST(REPLACE(VALOR_UNID, ',', '.') AS FLOAT) * UNIDADES) AS valor_total_gasto
FROM tb_vendas
ORDER BY valor_total_gasto DESC
LIMIT 5
"""

resultado_top5 = run_query(query_top_compras)
print(resultado_top5)
```

# 5) Faça uma consulta que retorne os produtos e a média do preço da unidade dos produtos, ordenando do maior para o menor.

```python
query_media_precos = """
SELECT
    PRODUTO,
    AVG(CAST(REPLACE(VALOR_UNID, ',', '.') AS FLOAT)) AS media_preco_unidade
FROM tb_vendas
GROUP BY PRODUTO
ORDER BY media_preco_unidade DESC
"""

df_medias = run_query(query_media_precos)
print(df_medias)
```

# 6) Faça uma consulta que retorne o id dos 3 clientes da base de vendas que mais relalizaram compras e a quantidade de compras realizadas.

```python
query_top_clientes = """
SELECT
    ID_CLIENTE,
    COUNT(ID_COMPRA) AS quantidade_compras
FROM tb_vendas
GROUP BY ID_CLIENTE
ORDER BY quantidade_compras DESC
LIMIT 3
"""

top_clientes = run_query(query_top_clientes)
print(top_clientes)
```
